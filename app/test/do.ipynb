```python
def run_llm_service(question, service_id="get_sparql_by_rag_service"):
    body = {
        "service_id": service_id,
        "params": {"input": question}
    }
    server_url = "http://127.0.0.1:8080/service_run"
    headers = { "Content-Type": "application/json" }
    response = requests.post(server_url, json=body, headers=headers)
    if response.status_code == 200:
        if response.json()["success"]:
            return json.loads(response.json()["result"])['output']
        else:
            raise Exception(f'failed to call service {service_id}.',
                            response.json()["message"])
    else:
        raise Exception(f'failed to call service {service_id}.',
                        response.status_code, response.text)
```

```python
SYS_PROMPT = """你是一个能够理解自然语言并将其转化为SPARQL查询的专业助手。你的任务是接收用户的提问，并生成一个正确的SPARQL查询语句。
- **输入格式**：用户的问题将以自然语言的形式给出，例如：“谁是《湖上草》的主要作者？”
- **输出格式**：你应当返回一个SPARQL查询语句，用于在相关知识图谱中查找答案。例如：“select ?x where { ?x <主要作品> <湖上草>. }”
- **关键点**：+
  - 确保SPARQL查询正确地反映了用户的问题。
  - 使用标准的SPARQL语法和命名空间。
  - 如果问题涉及多个实体或关系，请构建相应的多条件查询语句。
  - 回答简洁明了，直接提供SPARQL查询语句，无需额外的解释或说明。"""
```

```python
def get_sparql_from_llm(question):
    messages = [
        {"role": "system", "content": SYS_PROMPT},
        {"role": "user", "content": question}
    ]
    payload = {
        "model": "qwen2",
        "messages": messages,
        "temperature": 0.25,
        "top_p": 1,
        "stream": False,  # 是否开启流式输出，默认为False"
        "max_tokens": 128  # 根据需要调整回复的最大长度
    }
    # 发起POST请求
    response = requests.post(URL, headers=headers, json=payload)
    # 检查请求是否成功
    if response.status_code == 200:
        sparql = response.json()["choices"][0]["message"]["content"]
        # print(answer)
        return sparql, True
    else:
        return response.text, False
```

```python
def get_sparql_text(query=None, s=None, p=None, o=None, all=False):
    if query:
        query = re.sub(r' <[^>]*> <[^>]*> <[^>]*>(\.)', '', query)
        query = re.sub(r' <[^>]*> <[^>]*> <[^>]*> (\.)', '', query)
        query = re.sub(r' <[^>]*> <[^>]*> "[^>]*" (\.)', '', query)
        # query = re.sub(r' <[^>]*> <[^>]*> <[^>]*> .', '', query)
        # query = re.sub(r' <[^>]*> <[^>]*> <[^>]*>(\.\t)', '', query)
    elif s is not None and p is not None and o is not None:
        raise Exception("Invalid query")
    elif s is not None and p is not None:
        query = "select ?x where { <%s> <%s> ?x }" % (s, p)
    elif s is not None and o is not None:
        query = "select ?x where { <%s> ?x <%s> }" % (s, o)
    elif p is not None and o is not None:
        query = "select ?x where { ?x <%s> <%s> }" % (p, o)  
    elif s is not None:
        query = "select ?x ?y where { <%s> ?x ?y }" % (s)
    elif p is not None:
        query = "select ?x ?y where { ?x <%s> ?y }" % (p)
    elif o is not None:
        query = "select ?x ?y where { ?x ?y <%s> }" % (o)
    elif all:
        query = "select ?x ?y ?z where { ?x ?y ?z }"
    else:
        raise Exception("Invalid query")
    return query
```

```python
def get_answer_from_gStore(query):
    sparql_query = get_sparql_text(query)
    try:
        result = gc.query("ccks2024", "json", sparql_query, "POST").json()
        if result['StatusCode']==0:
            vars = result['head']['vars']
            if len(vars)==1:
                values = []
                for binding in result['results']['bindings']:
                    if binding[vars[0]]['type']=='uri':
                        values.append("<"+binding[vars[0]]['value']+">")
                    elif binding[vars[0]]['type'] in ['literal','typed-literal']:
                        values.append('"'+binding[vars[0]]['value']+'"')
                    else:
                        print(result)
                        raise Exception("Unknown type")
                answer = "\t".join(values)
                # 没有找到answer，重新获取sparql，再次查询
                if len(answer) > 0:
                    if len(answer) >2048:
                        answer = answer[:2048] + "..."
                    return answer, True
                else:
                    return "", False
            elif len(vars)==0:
                return "error query", False
            elif len(vars)>1:
                values = []
                for binding in result['results']['bindings']:
                    values.append([binding[var]['value'] for var in vars])
                print(values)
                return values, True
                # answer = "\n".join(values)
                # print(answer)
        else:
            print("error query:",result)
            return "error query", False
    except:
        raise Exception("error query")
    
```

```python

# 定义一个线程要执行的函数
def worker(task_queue):
    while True:
        task = task_queue.get()
        if task is None:
            break
        # get_sparql(task)
        get_answer(task)

# def upgrade_worker(task_queue):
#     while True:
#         data = upgrade_queue.get()
#         if data is None:
#             break
#         dataset[data.id] = data
#         print(f"Executing upgrade task {task}")

# 创建一个任务队列
task_queue = queue.Queue()
# upgrade_queue = queue.Queue()

# 创建线程
threads = []
for i in range(1):
    thread = threading.Thread(target=worker, args=(task_queue,))
    threads.append(thread)
    thread.start()

# upgrade_thread = threading.Thread(target=upgrade_worker, args=(upgrade_queue,))
# upgrade_thread.start()

# 添加任务到队列
# for i in range(5):
for i in dataset.values():
    if 'sparql' not in i:
        i['sparql'] = i['answer']
        del(i['answer'])
        
    elif 'answer' in i or 'log' in i:
        continue
    task_queue.put(i)
    # break

# 等待所有任务完成
for thread in threads:
    task_queue.put(None)
for thread in threads:
    thread.join()

# upgrade_queue.put(None)
# upgrade_thread.join()


print("All tasks are completed.")
```
